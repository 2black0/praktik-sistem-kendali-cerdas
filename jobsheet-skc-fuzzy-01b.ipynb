# Example MF

```python
x = np.arange(91)
x
```

```python
mf_cold   = fuzz.trimf(x, [0,0,45])
mf_normal = fuzz.trimf(x, [0,45,90])
mf_hot    = fuzz.trimf(x, [45,90,90])

xmf = [mf_cold, mf_normal, mf_hot]
label = ['cold', 'normal', 'hot']
```

```python
for xmfs, labels in zip(xmf, label):
    plt.plot(x, xmfs, label=labels)
    
plt.legend(loc="center")
plt.show()
```

```python
temp = 25

z_val = []
for i in range(len(xmf)):
    z = fuzz.interp_membership(x, xmf[i], temp)
    z_val.append(z)
    
print("_cold(25)=", z_val[0])
print("_normal(25)=", z_val[1])
print("_hot(25)=", z_val[2])
```

# AutoMF SKFuzzy

```python
from skfuzzy import control as ctrl
```

```python
x = np.arange(101)
x
```

```python
xmf = ctrl.Antecedent(x, 'xmf')
```

```python
names = ['sangat dingin' , 'dingin', 'agak dingin', 'normal', 'agak panas', 'panas', 'sangat panas']
```

```python
xmf.automf(names=names)
xmf.view()
```

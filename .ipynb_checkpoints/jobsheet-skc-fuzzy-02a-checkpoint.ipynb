# Fuzzy Set Operation

````python
import numpy as np
import matplotlib.pyplot as plt
````

````python
def tri_mf(x,a,b,c):
    res=[]
    for i in x:
        if i<a or i>c:
            res.append(0)
        if a<=i<b:
            res.append((i-a)/(b-a))
        if b<i<=c:
            res.append((c-i)/(c-b))
        if i==b:
            res.append(1)
    return res
````

````python
x_qual = np.arange(0, 11, 0.1)
````

````python
qual_lo = tri_mf(x_qual, 0, 0, 5);
````

````python
plt.plot(qual_lo);
````

````python
qual_md = tri_mf(x_qual, 0, 5, 10)
````

````python
plt.plot(qual_md);
````

````python
plt.plot(x_qual, qual_lo, x_qual, qual_md);
````

# Union

Consider 2 Fuzzy Sets denoted by A and  B, then let’s consider Y be the Union of them, then for every member of  A and  B, Y will be:|

```
degree_of_membership(Y)= max(degree_of_membership(A), degree_of_membership(B)) 
```

````python
def unionOp(x, xmf, y, ymf):
  z = []
  if len(x) != len(y):
    return "length of fuzzy set is not equal"
  else:
    for i in range(len(x)):
      if x[i] > y[i]:
        z.append(x[i])
      else:
        z.append(y[i])
  return z
````

````python
or_op = unionOp(qual_lo, x_qual, qual_md, x_qual)
````

````python
plt.plot(x_qual, qual_lo, x_qual, qual_md);
plt.plot(x_qual, or_op, linewidth=5);
````

# Intersection

Consider 2 Fuzzy Sets denoted by A and  B, then let’s consider Y be the Intersection of them, then for every member of  A and  B, Y will be:



```
degree_of_membership(Y)= min(degree_of_membership(A), degree_of_membership(B)) 
```

````python
def intersectionOp(x, xmf, y, ymf):
  z = []
  if len(x) != len(y):
    return "length of fuzzy set is not equal"
  else:
    for i in range(len(x)):
      if x[i] < y[i]:
        z.append(x[i])
      else:
        z.append(y[i])
  return z
````

````python
and_op = intersectionOp(qual_lo, x_qual, qual_md, x_qual)
````

````python
plt.plot(x_qual, qual_lo, x_qual, qual_md);
plt.plot(x_qual, and_op, linewidth=5);
````

# Complement

Consider a Fuzzy Sets denoted by A  , then let’s consider Y be the Complement of it, then for every member of  A  , Y will be:

```
degree_of_membership(Y)= 1 - degree_of_membership(A)
```

````python
def complementOp(x):
  z = []
  for i in range(len(x)):
    z.append(1-x[i])
  return z
````

````python
not_op = complementOp(qual_lo)
````

````python
plt.plot(x_qual, qual_lo);
plt.plot(x_qual, not_op, linewidth=5);
````
